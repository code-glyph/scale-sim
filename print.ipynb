```python
import math

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
x = [3.585, 4, 5, 6, 7, 8]
array_size = [12, 16, 32, 64, 128, 256]
# for mobilenet
cycles_is_m = [8473737.0, 5573227.0, 2314306.0, 1146690.0, 674195.0, 438928.0]
cycles_ws_m = [6183209.0, 3799529.0, 1433453.0, 672421.0, 395105.0, 287925.0]
cycles_os_m = [5943179.0,3682863.0,1338519.0,590421.0,334435.0,226896.0]

cycles_is_r = [14165393.0, 8539239.0, 2838997.0, 1165349.0, 595543.0, 332923.0]
cycles_os_r = [11972697.0, 6359601.0, 1718353.0, 547249.0, 262370.0, 157939.0]
cycles_ws_r = [13714819.0,7885563.0,2519815.0,910115.0,417628.0,223181.0]

def normalize(data):
    max_val = max(data)
    return [val / max_val for val in data]

def return_area_delay(data):
    return [a*b*b for a,b in zip(data,array_size)]

cycles_is_norm_m = normalize(return_area_delay(cycles_is_m))
cycles_os_norm_m = normalize(return_area_delay(cycles_os_m))
cycles_ws_norm_m = normalize(return_area_delay(cycles_ws_m))

cycles_is_norm_r = normalize(return_area_delay(cycles_is_r))
cycles_os_norm_r = normalize(return_area_delay(cycles_os_r))
cycles_ws_norm_r = normalize(return_area_delay(cycles_ws_r))

axs[0].plot(x, cycles_is_norm_m, color='blue', label="IS", marker='o')
axs[0].plot(x, cycles_os_norm_m, color='green', label="OS", marker='o')
axs[0].plot(x, cycles_ws_norm_m, color='red', label="WS", marker='o')
axs[0].set_title('mobilenet')
axs[0].set_xlabel('size')
axs[0].legend()
axs[0].set_ylabel('(relative) total cycles')
axs[0].grid(True)

axs[1].plot(x, cycles_is_norm_r, color='blue', label="IS", marker='o')
axs[1].plot(x, cycles_os_norm_r, color='green', label="OS", marker='o')
axs[1].plot(x, cycles_ws_norm_r, color='red', label="WS", marker='o')
axs[1].set_title('resnet18')
axs[1].set_xlabel('size')
axs[1].legend()
axs[1].set_ylabel('(relative) total cycles')
axs[1].grid(True)

fig.suptitle('area delay vs size of array', fontsize=16)
plt.tight_layout()  # Adjust the layout to fit title and labels
plt.show()
```
